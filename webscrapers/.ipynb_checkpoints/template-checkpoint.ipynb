```python
def create_download_path(wd, current_title, historical_title, victim_name, issue_date = False):
    if issue_date == False:
        final_download = f'{download_path }\\{current_title}\\{victim_name}\\{historical_title}'
    else:
        final_download = f'{download_path }\\{current_title}\\{victim_name}\\{historical_title}\\{issue_date}'
    return(final_download)
    


def browser_instance(executable_path, headless, download_path):
    
    
    chrome_options =  Options()
    if headless:
        chrome_options.add_argument("--headless")   
    chrome_options.add_experimental_option("prefs", {
            "download.default_directory": download_path})    
    driver = webdriver.Chrome(executable_path=executable_path, options = chrome_options)
    return driver    

def search_lynching(first_name, last_name, state, lynch_date, days_delta):
    # Get dates day_delta before and after lynching
    date_before_lynching = datetime.strptime(lynch_date,"%Y-%m-%d").date() - timedelta(days=7)
    date_after_lynching = datetime.strptime(lynch_date,"%Y-%m-%d").date() + timedelta(days=7)
    
    url = f'https://www.newspapers.com/search/#query={first_name}+{last_name}&p_province=us-{state}&ymd-start={date_before_lynching}&ymd-end={date_after_lynching}'
    return url

def get_date_from_url(url):      
    
    url2 = url.split("/")
    url3 = url2[8:11]
    url4 = "_".join(url3)
    return(url4)


def url_span(url, year, month):
    
    #month = int(month)
    #year = int(year)
    
    
    if month == 1:
        # Define time criteria
        month_before = 12       
        year_before = year - 1
        month_after = 2
        
        
        # Define variables to return
        link_before = f'{url}/{str(year_before)}/{str(month_before).zfill(2)}'
        link_current = f'{url}/{str(year)}/{str(month).zfill(2)}'
        link_after = f'{url}/{str(year)}/{str(month_after).zfill(2)}'
        
        
        
        
    elif month == 12:
        # Define time criteria
        month_after =1        
        year_after= year +1
        month_before = 11
        
        
        #Define variables to return
        link_before = f'{url}/{str(year)}/{str(month_before).zfill(2)}'
        link_current = f'{url}/{str(year)}/{str(month).zfill(2)}'
        link_after = f'{url}/{str(year_after)}/{str(month_after).zfill(2)}'
    
    else:
        # Define time criteria
        month_before = month -1
        month_after = month +1              
    
        # Define variables to return
        link_before = f'{url}/{str(year)}/{str(month_before).zfill(2)}'
        link_current = f'{url}/{str(year)}/{str(month).zfill(2)}'
        link_after = f'{url}/{str(year)}/{str(month_after).zfill(2)}'
        
   
     
    return link_before, link_current, link_after
    
    
    


def signIn(driver, url, signin_field, username_field, password_field):  
    driver.get(url)
    sign_in = driver.find_element_by_xpath(signin_field)
    sign_in.click()
    email_field = driver.find_element_by_xpath(username_field)
    password_field = driver.find_element_by_xpath(password_field)
    email_field.send_keys(USERNAME)
    password_field.send_keys(PASSWORD + Keys.RETURN)


def value_from_url(url):
    final_url = url[:-1].rsplit('/', 1)[-1]
    return final_url

def write_json_data(data_out, database_name, file_path):
        #writing data to json
        data_out = data_out
        database_name = database_name
        file = f'{file_path}/{database_name}'
        with open(file, 'w') as outfile:
            json.dump(data_out, outfile)
            
def get_element(driver,route,word = "NA"):   
    if word != "NA":
        complete_route = f'//*[contains(text(), "{word}")]/' + route    
    else:
        complete_route = route
    element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, complete_route)))
    
    return element

def get_list_column_entries(driver,url, word, column_route, full = True, duplicate_links = False):
    if url != "NA":
        driver.get(url)
    element_list = get_element(driver, column_route, word)
    element_list_url = element_list.find_elements_by_tag_name("a")
    list_elements = []
    counter = 1
    for element in element_list_url:
        if duplicate_links:
            if counter % 2 == 1: #there are two of same links, only need to grab one of them
                if full:
                    list_elements.append(element.get_attribute("href"))
                else:
                    list_elements.append(value_from_url(element.get_attribute("href")))
            counter = counter +1
        else:            
            if full:
                list_elements.append(element.get_attribute("href"))
            else:
                list_elements.append(value_from_url(element.get_attribute("href")))
    return(list_elements)

def get_person_results(results):
    newspaper_info = []
    for chunk in results:
        url = chunk.find_element_by_tag_name('a').get_attribute("href")
        newspaper_title = chunk.find_element_by_tag_name('h2').text
        other_information = chunk.find_elements_by_tag_name("p")    
        page_location = other_information[0].text
        city_state = other_information[1].text
        date = other_information[2].text
        #num_matches = other_information[3].text
        match = {"url": url, "newspaper_title": newspaper_title, "page_location": page_location, "city_state":city_state, "date": date}
        newspaper_info.append(match)
    return newspaper_info

def get_ocr_text(newspaper_urls):
    """
    input: a list of newspaper urls that link to individual photo pages
    action: changes the photo url to a ocr url
    output: saves the OCR results from newspaper.com for each photo given
    """
    result_ocr = []
    for newspaper_url in newspaper_urls:        
        ocr_string = "https://www.newspapers.com/newspage/"
        ocr_url = ocr_string + newspaper_url[33:]
        print(ocr_url)
        driver.get(ocr_url)
        ocr_text = driver.find_element_by_class_name("ocrtext").text
        match_ocr = {"url": newspaper_url, "ocr_text": ocr_text}
        result_ocr.append(match_ocr)
    return result_ocr

def save_ocr_text(driver,newspaper_url, page_number, download_path ):
    """
    input: a list of newspaper urls that link to individual photo pages
    action: changes the photo url to a ocr url
    output: saves the OCR results from newspaper.com for each photo given
    """
    ocr_string = "https://www.newspapers.com/newspage/"
    ocr_url = ocr_string + newspaper_url[33:]
    print(ocr_url)
    driver.get(ocr_url)
    ocr_text = driver.find_element_by_class_name("ocrtext").text
    text_download_path = f'{download_path}\\test_page_{str(page_number)}.txt'
    file1 = open(text_download_path, "w")
    file1.write(ocr_text)
    file1.close()
    return("done")
   

def save_photo(driver, url, printSaveButton_route = printSaveButton_route, saveEntirePage_route= saveEntirePage_route, saveButton_route= saveButton_route, saveJpeg_route = saveJpeg_route ):
    driver.get(url)
    time.sleep(2)
    #Initial Save button
    print_save_button = get_element(driver, printSaveButton_route)        
    print_save_button.click()
    print("clicked")

    #Save Entire page
    entire_page_button = get_element(driver, saveEntirePage_route )
    entire_page_button.click()

    #Save page
    save_button = get_element(driver,saveButton_route)   
    save_button.click()

    #Save Jpeg
    save_jpeg_button = get_element(driver, saveJpeg_route)
    save_jpeg_button.click()
    time.sleep(7)
    
    #
```

```python
def get_newspaper_info(df):
    counter = 0
    # Initalize variables for error log
    ls_victim_name = []
    ls_current_title = []
    ls_historical_title = []
    ls_date = []
    ls_url = []
    ls_pages_expected = []
    ls_num_ocr = []
    ls_num_jpg = []
    ls_num_txt = []
    ls_txt_status = []
    ls_jpg_status = []
    error_log = {}
    for row in df.itertuples():
        counter = counter +1
        ls_url_span = url_span(row.unique_combined_urls,row.year,row.month ) # returns a tuple of the three urls for a timeframe
        # define variable names for folder names
        current_title = row.current_title.replace('.', '')
        historical_title = row.title.replace('.', '')
        victim_name = row.victim_name
        
        for url in ls_url_span[1:]:
            print(f'url: {url}')
            # These steps get to the date
           # (executable_path, headless, download_path, current_title, historical_title, victim_name):
            download_path_mid = create_download_path(download_path, current_title, historical_title, victim_name)
            print(f"download path mid: {download_path_mid}")
            driver = browser_instance(executable_path, False, download_path_mid)
            signIn(driver, newspaperdotcomURL, signin_field, username_field, password_field)
            time.sleep(3)                    
            
            # This gets the list of date urls
            ls_dates_scrape = get_list_column_entries(driver,url, "DATE", column_route)
            
            for date in ls_dates_scrape:
                print(f'date: {date}')
                ls_pages_scrape = get_list_column_entries(driver,date, "PAGE", column_route)
                num_pages = len(ls_pages_scrape)
                print(f'num pages: {num_pages}')
                page_number = 0
                
                # quite old driver session and spin up a new one with new folder download path that has the date range necessary              
                driver.quit()
                issue_date = get_date_from_url(date)
                download_path_final = create_download_path(download_path, current_title, historical_title, victim_name, issue_date)
                driver =  browser_instance(executable_path, False, download_path_final)
                signIn(driver, newspaperdotcomURL, signin_field, username_field, password_field)
                time.sleep(3)
                for page in ls_pages_scrape:
                    page_number = page_number +1
                    print(f'page: {page}')
                    save_photo(driver, page)
                    save_ocr_text(driver, page, page_number, download_path_final)
            # Check to see if successful
            
                ls_victim_name.append(victim_name)
                ls_current_title.append(current_title)
                ls_historical_title.append(historical_title)
                ls_url.append(url)
                ls_date.append(date)
                ls_pages_expected.append(num_pages)
                num_txt = len(fnmatch.filter(os.listdir(download_path_final), '*.txt'))
                num_jpg = len(fnmatch.filter(os.listdir(download_path_final), '*.jpg'))
                ls_num_txt.append(num_txt)
                ls_num_jpg.append(num_jpg)
                if num_txt == num_pages:
                    ls_txt_status.append(1)
                if num_txt != num_pages:
                    ls_txt_status.append(0)
                if num_jpg == num_pages:
                    ls_jpg_status.append(1)
                if num_jpg != num_pages:
                    ls_jpg_status.append(0)
                data = {"victim_name": ls_victim_name, "current_title": ls_current_title, "historical_title": ls_historical_title, "url":ls_url, "date":ls_date, "pages_expected": ls_pages_expected, "num_txt":ls_num_txt, "num_jpg":ls_num_jpg, "jpg_status": ls_jpg_status, "txt_status": ls_txt_status}    
                df = pd.DataFrame(data)
                df.to_csv("error_log.csv")
            
            
                
        # for test purposes
        if counter > 0:
            break
    print("end-----")
```
